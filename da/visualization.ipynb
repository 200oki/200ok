```python
from os import path
import os
from pprint import pprint


import pandas as pd
import numpy as np
```

```python
characters = pd.read_excel('./datasets/characters.xlsx')
characters
```

## personality

```python
characters.groupby('personality').size().reset_index(name='personality_counts')
```

# Hobby

```python
characters.groupby('hobby').size().reset_index(name='hobby_counts')
```

# Birthday

```python
# 생일 같은 주민이 있는지 확인 => 존재 (최대 3명까지 겹침. 06-07)
birth = characters.groupby('birthday').size().reset_index(name='counts')
birth
# birth.loc[birth['counts'].idxmax()]

# 생일이 없는 날이 있는지 확인 => 없는 날 04-01
```

## Birthday MM
생일별 달 분포

```python
birth.birthday.str.split('-').str[0].reset_index(name='month')
birth.groupby('month').size().reset_index(name='month_counts')
```

## Birthday Day
생일별 일 분포 <br>

```python
# day 비교 => 01 은 11개 04-01이 없음 / 31은 7개 1, 3, 5, 7, 8, 10, 12
birth.groupby(birth.birthday.str.split('-').str[1]).size().reset_index(name='day_counts')
```
