## 주민 인기도 순위 데이터에 이름을 붙입니다.

방법은 주민 데이터와 같습니다.


```python
import numpy as np
import pandas as pd


CSVPATH = './datasets/acnh-pop/acnh_villager_data.csv'
TRANSPATH = './datasets/koname.txt'
```

```python
data = pd.read_csv(CSVPATH)
trans_csv = pd.read_csv(TRANSPATH, delimiter='\t')
trans = dict(zip(trans_csv['en'], trans_csv['ko']))
```

```python
data.info()
ennames = data['name'].copy()
data['name'] = data['name'].apply(lambda v: trans[v] if v in trans else v)
print()
data.info()
```

413 개 레코드 중에 21개나 비었습니다.

```python
# 마을 주민 데이터는 전부 Name으로 돼 있습니다.
# data.rename(columns={'name': 'Name'})

# 어느 녀석들이 비었는지 봅니다.
# data[data['name'].isnull()]
data[data['name'].str.match(r'^[a-zA-Z]')]
```

몇개 안되니까 인터넷에서 찾아서 고치도록 하겠습니다.

```python
data['name'].replace(
    {
        'Shino': '요비',
        'Sasha': '미첼',
        'Ione': '스피카',
        'Cephalobot': '기가',
        'Tiansheng': '제천',
        'Petri': '리카',
        'Quinn': '시온',
        'Marlo': '보스터',
        'Azalea': '페튜니아',
        'Faith': '마치',
        'Chabwick': '펭구',
        'WartJr': '샘',
        'Rio': '데자네',
        'Buck(Brows)': '바야시코프',
        'Renee': '뿔님이',
        'Roswell': '삐로코',
        'Zoe': '소면',
        'OHare': '산토스',
        'Crackle(Spork)': '포크',
        'Ace': '페더',
        'Frett': '샹펜',
    },
    inplace=True
)
ennames.replace(
    {
        'WartJr': 'Wart Jr.',
        'Buck(Brows)': 'Buck',
        'Renee': 'Renée',
        'OHare': "O'Hare",
    },
    inplace=True
)

data[data['name'].str.match(r'^[a-zA-Z]')]
```
